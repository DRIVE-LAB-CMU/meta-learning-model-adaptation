```python
# MPPI_profiling
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = json.load(open('/home/lecar/car/car_collect/data/dataset/data-20240327-233254/time.json'))
# plt.plot(data['total'][1:])
plt.plot(data['mppi'][1:], label='mppi total')
plt.plot(data['mppi_sample'][1:], label='sample')
plt.plot(data['mppi_rollout'][1:], label='rollout')
plt.plot(data['mppi_misc'][1:], label='misc')
plt.plot(data['mppi_reward'][1:], label='reward')
plt.legend()
plt.title('4090')
plt.ylim([0, 0.009])
print(np.mean(data['mppi_rollout'][1:]))
```
