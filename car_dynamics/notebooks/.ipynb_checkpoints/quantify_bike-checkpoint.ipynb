```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# import utm
import scipy
import scipy.interpolate
import scipy.signal
from scipy.spatial.transform import Rotation
from envs.kinematic_bicycle_model import LF, LR
import ipywidgets as widgets
from ipywidgets import interact
from termcolor import colored
%matplotlib inline
```

```python
def pos2vel_savgol(p, delta=0.05, window_length=9, polyorder=2):
    assert len(p.shape) > 1
    v_smooth = np.zeros_like(p)
    p_smooth = np.zeros_like(p)
    a_smooth = np.zeros_like(p)
    for i in range(p.shape[1]):
            p_smooth[:,i] = scipy.signal.savgol_filter(p[:,i], window_length=window_length, polyorder=polyorder, \
                                                       deriv=0, delta=delta)
            v_smooth[:,i] = scipy.signal.savgol_filter(p[:,i], window_length=window_length, polyorder=polyorder, \
                                                       deriv=1, delta=delta)
            a_smooth[:,i] = scipy.signal.savgol_filter(p[:,i], window_length=window_length, polyorder=polyorder, \
                                                       deriv=2, delta=delta)
    
    # kappa = np.abs(v_smooth[:,0]*a_smooth[:,1]-v_smooth[:,1]*a_smooth[:,0]) /\
    #                     (v_smooth[:,0]**2. + v_smooth[:,0]**2.) ** 1.5

    v_smooth = v_smooth[:,:2]
    v_smooth = np.linalg.norm(v_smooth, axis=1)
    # return p_smooth, v_smooth, kappa
    return p_smooth, v_smooth

def calc_delta(state, ):
    '''
    Return
        delta, beta
    '''
    delta = []
    alpha_list = []
    for i, _ in enumerate(state):
        if i == 0:
            continue
        ld = np.linalg.norm(state[i,:2] - state[i-1,:2])
        
        # R = 1 / kappa[i-1]
        
        car_dir = np.array([np.cos(state[i-1,2]),np.sin(state[i-1,2])])
        vel_dir = np.array(state[i,:2] - state[i-1,:2])
        alpha = np.math.atan2(np.linalg.det([car_dir, vel_dir]),np.dot(car_dir,vel_dir))
        
        alpha_list.append(alpha)
        # alpha = np.arcsin(ld / R)
        # print(R)
        delta.append(np.arctan2(np.tan(alpha)*(LF+LR),LR))
    # return np.array(delta)
    return np.array(delta), np.array(alpha_list)
```

```python
def interpolation(t, x, frequency=20, mode='linear', t_new=None):
    assert len(x.shape) > 1
    # t: scalar
    if t_new is None:
        l = int(np.floor((t[-1]-t[0])*frequency)) + 1
        t_new = np.linspace(t[0], t[0] + (l-1)/frequency, l) 
    else:
        l = len(t_new)
    
    x_new = np.zeros((l, x.shape[1]))
    for i in range(x.shape[1]):
        f = scipy.interpolate.interp1d(t, x[:,i], kind=mode, bounds_error=False, fill_value='extrapolate')
        try:
            x_new[:,i] = f(t_new)
        except:
            print(t.shape, x[:,i].shape, t_new.shape, x_new.shape)
    return t_new, x_new
```

```python
data_dir = './dataset/car-data-x/'

files = {
    'throw': [
        {'path':'2023-10-10-22_11_11-wqf.txt','st':5,'ed':-1},
    ],
    'drag': [
        {'path':'2023-10-10-22_19_29-ek4.txt','st':5,'ed':-1},
    ],
    'box': [
        {'path':'2023-10-10-22_15_14-hwd.txt','st':5,'ed':-1},
    ],
    'car': [
        {'path':'2023-10-10-21_35_06-pvg.txt','st':5,'ed':-1},
        {'path':'2023-10-10-21_38_31-572.txt','st':5,'ed':-1},
        
    ],
}

def rotate_180(angles):
    new_angles = np.zeros_like(angles)
    for i, angle in enumerate(angles):   
        xs = np.cos(angle)
        ys = np.sin(angle)
        new_angles[i] = np.arctan2(-ys,-xs)
    return new_angles

def calc_vel(data):
    vel_calc = [0.]
    for i in range(1,len(data)):
        dist = np.linalg.norm(data[i][:2] - data[i-1][:2])
        vel_calc.append(dist/(data[i,-1]-data[i-1,-1]))
    return np.array(vel_calc)
```

```python
def error_xy(state, vel, delta, abs=True):
    frequency = 20
    dt = 1./frequency
    # error x
    delta_x = state[1:, 0] - state[:-1,0]
    delta_x_hat = vel * np.cos(state[:-1,2] + delta) * dt
    error_x = (delta_x - delta_x_hat) / delta_x
    # error y
    delta_y = state[1:, 1] - state[:-1,1]
    delta_y_hat = vel * np.sin(state[:-1,2] + delta) * dt
    error_y = (delta_y - delta_y_hat) / delta_y
    if abs:
        error_x, error_y = np.abs(error_x), np.abs(error_y)
    return error_x, error_y
```

```python
data_dict = {}
for EXP_TYPE in ['car','box','drag','throw']:    
    f = files[EXP_TYPE][0]['path']
    st = files[EXP_TYPE][0]['st']
    ed = files[EXP_TYPE][0]['ed']
    data = np.loadtxt(os.path.join(data_dir,f))
    data = data[st:ed]
    data[:,2] = rotate_180(data[:,2])
    data[:,3] = calc_vel(data)
    data[:,-1] -= data[0,-1]
    # pos_x, pos_y, yaw, vel, throttle, str_val, _= data[i-1]
    data_dict[EXP_TYPE] = data
```

```python
def plot_input_to_target_error(EXP_TYPE, speed_max=5.5, steer_max=.4,delta_offset = 0.):
    '''Error between 
        - vel_real and vel_target
        - steer_real and steer_target          
    '''
    fig, axs = plt.subplots(1,2,figsize=(15,5))
    frequency = 20
    dt = 1./frequency
    data = data_dict[EXP_TYPE]
    t_20hz, state_20hz = interpolation(data[:,-1],data, frequency=frequency)
    state, vel = pos2vel_savgol(state_20hz,window_length=5)
    delta, beta = calc_delta(state)
    
    vel_real = vel[:-1]
    vel_tar = state[:-1,4]*speed_max
    delta_real = delta
    delta_tar = state[:-1,5]*steer_max + delta_offset
    t = t_20hz[:-1]

    axs[0].plot(vel_real,label='real')
    axs[0].plot(vel_tar,label='target')

    axs[1].plot(delta_real,label='real')
    axs[1].plot(delta_tar,label='target')
    axs[0].set_title("velocity")
    axs[1].set_title("steering")
    axs[0].legend()
    axs[1].legend()
    print(colored(f'MSE Vel:   {np.mean((vel_real - vel_tar)**2)}','green'))
    print(colored(f'MSE Steer: {np.mean((delta_real - delta_tar)**2)}','green'))
    
def plot_closed_loop_error(EXP_TYPE, speed_max=5.5, steer_max=.4,vis_y_max=5.):
    fig, axs = plt.subplots(1,2,figsize=(15,5))
    frequency = 20
    dt = 1./frequency
    data = data_dict[EXP_TYPE]
    t_20hz, state_20hz = interpolation(data[:,-1],data, frequency=frequency)
    state, vel = pos2vel_savgol(state_20hz,window_length=5)
    delta, beta = calc_delta(state,)
    
    vel_real = vel[:-1]
    vel_tar = state[:-1,4]*speed_max
    delta_real = beta
    delta_tar = np.array([np.arctan2(np.tan(state[i,5]*steer_max)*LR,(LR+LF)) \
                         for i in range(state.shape[0]-1)])
    
    
    t = t_20hz[:-1]

    error_x_list = []
    error_y_list = []
    
    error_x, error_y = error_xy(state, vel_real, delta_real)
    error_x_list.append(error_x)
    error_y_list.append(error_y)
    axs[0].plot(t, error_x,label='real vel; real delta', color='red')
    axs[1].plot(t, error_y,label='real vel; real delta', color='red')
    # plt.plot(error_y)
    # print(np.mean(error_x), np.mean(error_y))
    
    error_x, error_y = error_xy(state, vel_real, delta_tar)
    error_x_list.append(error_x)
    error_y_list.append(error_y)
    axs[0].plot(t, error_x,label='real vel; target delta')
    axs[1].plot(t, error_y,label='real vel; target delta')
    # plt.plot(error_y)
    
    error_x, error_y = error_xy(state, vel_tar, delta_real)
    error_x_list.append(error_x)
    error_y_list.append(error_y)
    axs[0].plot(t, error_x,label='target vel; real delta')
    axs[1].plot(t, error_y,label='target vel; real delta')
    # plt.plot(error_y)
    
    error_x, error_y = error_xy(state, vel_tar, delta_tar)
    error_x_list.append(error_x)
    error_y_list.append(error_y)
    axs[0].plot(t, error_x,label='target vel; target delta')
    axs[1].plot(t, error_y,label='target vel; target delta')
    # plt.plot(error_y)
    
    axs[0].set_ylabel('pos_x absolute error')
    axs[0].set_xlabel('t (s)')
    axs[1].set_ylabel('pos_y absolute error')
    axs[1].set_xlabel('t (s)')
    axs[0].legend()
    axs[0].set_yscale('log')
    axs[1].set_yscale('log')
    fig.suptitle(f"{EXP_TYPE}")

    plt.figure()
    # width of the bars
    barWidth = 0.3
    
    bars1 = [np.mean(error_x) for error_x in error_x_list]
     
    # Choose the height of the cyan bars
    bars2 = [np.mean(error_y) for error_y in error_y_list]
     
    # Choose the height of the error bars (bars1)
    yer1 = [np.std(error_x) for error_x in error_x_list]
     
    # Choose the height of the error bars (bars2)
    yer2 = [np.std(error_y) for error_y in error_y_list]
     
    # The x position of bars
    r1 = np.arange(len(bars1))
    r2 = [x + barWidth for x in r1]
     
    # Create blue bars
    plt.bar(r1, bars1, width = barWidth, color = 'blue', edgecolor = 'black', yerr=yer1, capsize=7, label='pos_x')
     
    # Create cyan bars
    plt.bar(r2, bars2, width = barWidth, color = 'cyan', edgecolor = 'black', yerr=yer2, capsize=7, label='pos_y')
     
    # general layout
    plt.xticks([r + barWidth for r in range(len(bars1))], ['vel; delta','vel; delta_tar','vel_tar;delta','vel_tar;delta_tar'])
    plt.ylabel('error')
    plt.yticks([0., 1.,2.],['0%','100%','200%'])
    plt.ylim([-1,vis_y_max])
    plt.legend()
    # plt.yscale('log')    


def plot_open_loop_trajectory(EXP_TYPE, real_yaw=False, vel_type='real vel', 
                              steer_type='real steer',speed_max=5.5,steer_max=.4,
                             beta_offset=0.1, delta_offset = 0.1):
    frequency = 20
    dt = 1./frequency
    data = data_dict[EXP_TYPE]
    t_20hz, state_20hz = interpolation(data[:,-1],data, frequency=frequency)
    state, vel = pos2vel_savgol(state_20hz,window_length=5)
    delta, beta = calc_delta(state)

    if vel_type != 'real vel':
        vel = state[:-1,4]*speed_max
    if steer_type != 'real steer':
        beta = np.array([np.arctan2(np.tan(state[i,5]*steer_max + delta_offset)*LR,(LR+LF)) \
                         for i in range(state.shape[0]-1)])
        delta = np.arctan2(np.tan(beta)*(LF+LR),LR)
    else:
        delta = delta - delta_offset
        beta = beta - beta_offset
        
    st=0
    x_list = []
    y_list = []

    delta_list = []
    x, y, psi = state[st,:3]
    
    x_list.append(x)
    y_list.append(y)
    
    # print(state[1:3]-state[0:2])
    for v_t, beta_t, delta_t, yaw_t in zip(vel[st:], beta[st:], delta[st:], state[st:,2]):
        dx = v_t * np.cos(psi + beta_t) * dt
        dy = v_t * np.sin(psi + beta_t) * dt
        # print(psi, yaw_t)
        # psi += v_t * np.sin(beta_t) / LR * dt
        if real_yaw == False:
            psi += v_t * np.cos(beta_t) * np.tan(delta_t) / (LR+LF) * dt
        else:
            psi = yaw_t
        # print(dx, dy)
        x += dx
        y += dy
        x_list.append(x)
        y_list.append(y)
        delta_list.append(delta_t)
        
    # vel_real = vel[:-1]
    # vel_tar = state[:-1,4]*speed_max
    # delta_real = beta
    # delta_tar = np.array([np.arctan2(np.tan(state[i,5]*steer_max)*LR,(LR+LF)) \
    #                      for i in range(state.shape[0]-1)])
    
    
    # t = t_20hz[:-1]
    plt.title(f"Trajectory: {EXP_TYPE}")
    plt.plot(x_list,y_list)
    plt.plot(state[:,0],state[:,1])
    plt.xlabel('x')
    plt.ylabel('y')
```

# Visualize error of the kinematic bicycle model (S-line data)

We assume $\delta_r=0$, then:

- $\dot x = v cos(\psi + \beta)$
- $\dot y = v sin(\psi + \beta)$
- $\dot \psi = \frac{v\cos(\beta)}{L} (\tan\delta_f - \tan\delta_r)=\frac{v\cos(\beta)}{L} \tan\delta_f$

then the steering angle
$$
\beta = \tan^{-1}\frac{L_f\tan\delta_r + L_r  \tan \delta_f}{L} = \tan^{-1}\frac{L_r \tan \delta_f}{L},
$$
$$
\delta_f = \tan^{-1} \frac{L\tan\beta}{L_r}
$$

Remark:
 - error of (real vel, real delta) use $\beta$ 0.06134403083526475 0.0613440308352648
 - error of (real vel, real delta) use $\delta_f$ 0.40519499427142325 0.19398476325222525

![](https://miro.medium.com/v2/resize:fit:640/format:webp/1*e7sQaBLkEwRSILi-NykBpw.png)

Based on the following analysis, we assume there are two types of mismatch:

1. Mismatch between control and real $\delta_f$, i.e., $\delta_f = a \times u + b$.
2. Measurement error between Mocap yaw and real yaw $\beta = \beta_{\text{calculated}} + e$
3. Slipping wheel $\delta_f = \delta_f(\text{calculated}) + b$
