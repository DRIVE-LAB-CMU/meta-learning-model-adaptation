```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import utm
import scipy
import scipy.interpolate
import scipy.signal
from scipy.spatial.transform import Rotation
import ipywidgets as widgets
%matplotlib inline
```

```python
def pos2vel_savgol(t, p, delta=0.05, window_length=9, polyorder=2):
    v_smooth = np.zeros_like(p)
    p_smooth = np.zeros_like(p)
    a_smooth = np.zeros_like(p)
    if len(p.shape) == 1:
        p_smooth = scipy.signal.savgol_filter(p, window_length=window_length, polyorder=polyorder, \
                                              deriv=0, delta=delta)
        v_smooth = scipy.signal.savgol_filter(p, window_length=window_length, polyorder=polyorder, \
                                              deriv=1, delta=delta)
        a_smooth = scipy.signal.savgol_filter(p, window_length=window_length, polyorder=polyorder, \
                                              deriv=2, delta=delta)
    else:
        for i in range(p.shape[1]):
            p_smooth[:,i] = scipy.signal.savgol_filter(p[:,i], window_length=window_length, polyorder=polyorder, \
                                                       deriv=0, delta=delta)
            v_smooth[:,i] = scipy.signal.savgol_filter(p[:,i], window_length=window_length, polyorder=polyorder, \
                                                       deriv=1, delta=delta)
            a_smooth[:,i] = scipy.signal.savgol_filter(p[:,i], window_length=window_length, polyorder=polyorder, \
                                                       deriv=2, delta=delta)
    return p_smooth, v_smooth, a_smooth
```

```python
def interpolation(t, x, frequency=20, mode='linear', t_new=None):
    # t: scalar
    if t_new is None:
        l = int(np.floor((t[-1]-t[0])*frequency)) + 1
        t_new = np.linspace(t[0], t[0] + (l-1)/frequency, l) 
    else:
        l = len(t_new)
    
    if len(x.shape) == 1:
        f = scipy.interpolate.interp1d(t, x, kind=mode, bounds_error=False, fill_value='extrapolate')
        x_new = f(t_new)
    else:
        x_new = np.zeros((l, x.shape[1]))
        for i in range(x.shape[1]):
            f = scipy.interpolate.interp1d(t, x[:,i], kind=mode, bounds_error=False, fill_value='extrapolate')
            try:
                x_new[:,i] = f(t_new)
            except:
                print(t.shape, x[:,i].shape, t_new.shape, x_new.shape)

    return t_new, x_new
```

```python
def get_data(filename, filename_stack, ELEV=False, mode='raw'):
    if mode == 'raw':
        data_pos = pd.read_csv(filename + '_slash_crl_rzr_slash_duro_slash_piksi_slash_pos_llh_cov.csv')
    else:
        data_pos = pd.read_csv(filename + '_slash_crl_rzr_slash_duro_slash_piksi_slash_pos_llh_cov_ins.csv')
    data_odometry = pd.read_csv(filename + '_slash_crl_rzr_slash_vehicle_slash_transmission_slash_odometry.csv')
    data_quat = pd.read_csv(filename + '_slash_crl_rzr_slash_duro_slash_piksi_slash_orient_quat.csv')
    data_vehicle = pd.read_csv(filename + '_slash_crl_rzr_slash_vehicle.csv')
    data_vehicle_control = pd.read_csv(filename + '_slash_crl_rzr_slash_vehicle_slash_control_status.csv')
    if filename_stack is not None:
        data_pid = pd.read_csv(filename_stack + '_slash_crl_rzr_slash_pid_info.csv')
        data_input = pd.read_csv(filename_stack + '_slash_crl_rzr_slash_input.csv')
        if ELEV:
            data_elev = pd.read_csv(filename_stack + '_slash_crl_rzr_slash_elevation_map.csv')
    
    Data = {}
    Data['time'] = 1e-9 * (np.copy(data_pos.iloc[:,0]) - data_pos.iloc[0,0])

    gps_raw = np.copy(data_pos.iloc[:,10:13])
    pos_raw = np.zeros_like(gps_raw)
    pos_raw[:,0], pos_raw[:,1] = utm.from_latlon(gps_raw[:,0], gps_raw[:,1])[0:2]
    pos_raw[:,2] = gps_raw[:,2]
    
    for i in range(3):
        pos_raw[:,i] -= pos_raw[0,i]
    
    Data['pos'] = pos_raw
    
    # interpolation
    Data['time'], Data['pos'] = interpolation(Data['time'], Data['pos'], frequency=20)

    # compute velocity
    Data['pos_smooth'], Data['vel'], Data['acc'] = pos2vel_savgol(Data['time'], Data['pos'], delta=0.05, window_length=21, polyorder=2)
    Data['lin_vel'] = np.sqrt(Data['vel'][:,0]**2 + Data['vel'][:,1]**2)
    
    # get linear velocity from odometry
    t = 1e-9 * (np.copy(data_odometry.iloc[:,0]) - data_odometry.iloc[0,0])
    lin_vel_odom = np.copy(data_odometry.iloc[:,23])
    _, Data['lin_vel_odom'] = interpolation(t, lin_vel_odom, t_new=Data['time'], frequency=20)
    
    # get quat and euler
    t = 1e-9 * (np.copy(data_quat.iloc[:,0]) - data_quat.iloc[0,0])
    quat = np.copy(data_quat.iloc[:,8:12])
    _, Data['quat'] = interpolation(t, quat, t_new=Data['time'], frequency=20)
    Data['euler_zyx'] = np.zeros_like(Data['pos'])
    for i in range(len(Data['time'])):
        Data['quat'][i,:] = Data['quat'][i,:] / np.linalg.norm(Data['quat'][i,:])
        Data['euler_zyx'][i,:] = Rotation.from_quat(Data['quat'][i,:]).as_euler('ZYX')
    
    # get steer angle 
    t = 1e-9 * (np.copy(data_vehicle.iloc[:,0]) - data_vehicle.iloc[0,0])
    steer = np.copy(data_vehicle.iloc[:,24])
    _, Data['steer'] = interpolation(t, steer, t_new=Data['time'], frequency=20)
    
    # get steer percentage 
    t = 1e-9 * (np.copy(data_vehicle_control.iloc[:,0]) - data_vehicle_control.iloc[0,0])
    steer_percentage = np.copy(data_vehicle_control.iloc[:,11])
    _, Data['steer_percentage'] = interpolation(t, steer_percentage, t_new=Data['time'], frequency=20)
    
    # compute radius
    Data['radius'] = (Data['vel'][:,0]**2 + Data['vel'][:,1]**2)**1.5 / \
                      np.abs(Data['acc'][:,0]*Data['vel'][:,1] - Data['acc'][:,1]*Data['vel'][:,0])
    
    # compute normal acceleration
    Data['normal_acc'] = Data['lin_vel']**2 / Data['radius'] 
    
    # compute normal acceleration directly from steer angle 
    Data['normal_acc_steer'] = Data['lin_vel']**2 * np.sin(np.abs(Data['steer_percentage'])*0.54) / 4
    
    # elevation map
    if (filename_stack is not None) and ELEV:
        t = 1e-9 * (np.copy(data_elev.iloc[:,0]) - data_elev.iloc[0,0])
        elev_raw = data_elev.iloc[:,24]
        elev = np.zeros((len(t), 16384))
        for i in range(len(t)):
            elev[i, :] = np.fromstring(elev_raw[i][1:-1], sep=',')
        _, elev_interp = interpolation(t, elev, t_new=Data['time'], frequency=20)
        Data['elev'] = np.zeros((len(Data['time']), 128, 128))
        for i in range(len(Data['time'])):
            Data['elev'][i, :, :] = elev_interp[i, :].reshape((128, 128))
    
    # PID statistics
    if filename_stack is not None:
        t = 1e-9 * (np.copy(data_pid.iloc[:,0]) - data_pid.iloc[0,0])
        _, Data['pid_vd'] = interpolation(t, np.copy(data_pid.iloc[:,1]), t_new=Data['time'], frequency=20)
        _, Data['pid_v'] = interpolation(t, np.copy(data_pid.iloc[:,2]), t_new=Data['time'], frequency=20)
        Data['pid_error'] = Data['pid_v'] - Data['pid_vd']
    
    # Brake and throttle
    if filename_stack is not None:
        t = 1e-9 * (np.copy(data_input.iloc[:,0]) - data_input.iloc[0,0])
        _, Data['throttle'] = interpolation(t, np.copy(data_input.iloc[:,2]), t_new=Data['time'], frequency=20)
        _, Data['brake'] = interpolation(t, np.copy(data_input.iloc[:,3]), t_new=Data['time'], frequency=20)
    
    # compute 3-d rollover risk
    Data['3d_rollover_left'] = Data['lin_vel']**2 / Data['radius'] + 9.81 * np.sin(Data['euler_zyx'][:,2])
    Data['3d_rollover_right'] = Data['lin_vel']**2 / Data['radius'] - 9.81 * np.sin(Data['euler_zyx'][:,2])
    
    return Data
```
